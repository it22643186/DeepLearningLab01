```python

print("--- Exponential Distribution ---")
exponential_array_4x4 = np.random.exponential(scale=6.0, size=(100000))
print("array from exponential distribution:\n", exponential_array_4x4)
```

```python
import numpy as np
import matplotlib.pyplot as plt

uniform_data = np.random.rand(100000)
normal_data = np.random.randn(100000)


plt.figure(figsize=(10, 6)) 
plt.hist(uniform_data, density=True,bins=100, histtype="step", color="blue", label="Uniform (rand)")
plt.hist(normal_data, density=True, bins=150, histtype="step", color="green", label="Normal (randn)")
plt.hist(exponential_array_4x4, density=True, bins=150, histtype="step", color="red", label="Exponential")

plt.axis([-2.5, 2, 0, 1.1]) 
plt.legend(loc="upper right") 
plt.title("Comparison of Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True, linestyle='--', alpha=0.7) 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D # Required for 3D plotting

# --- 3D Plot of Z = X^2 + Y^2 ---
print("3D Plot of Z = X^2 + Y^2 ---")

# Create a figure and a 3D axes object
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Define the range for X and Y values
x = np.linspace(-5, 5, 100) # 100 points between -5 and 5
y = np.linspace(-5, 5, 100) # 100 points between -5 and 5


X, Y = np.meshgrid(x, y)

# Calculate Z based on the formula Z = X^2 + Y^2
Z = X**2 + Y**2

# Plot the surface.
# cmap='viridis' sets the color map for the surface.
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='red')

# Set labels for the axes
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis (X^2 + Y^2)')
ax.set_title('3D Plot of Z = X^2 + Y^2')

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 

print(" Pearson and Spearman Correlation Heatmaps ---")

df = pd.read_csv('Pokemon.csv',encoding='cp1252')
features_for_correlation = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
df_selected_features = df[features_for_correlation]

pearson_corr = df_selected_features.corr(method='pearson')
print("Pearson Correlation Matrix:\n", pearson_corr)

plt.figure(figsize=(9, 7)) 
sns.heatmap(pearson_corr,
            annot=True,    
            cmap='coolwarm', 
            fmt=".2f",     
            linewidths=.5, 
            cbar_kws={'label': 'Pearson Correlation Coefficient'}) 
plt.title('Pearson Correlation Heatmap of Pokémon Features')
plt.show()
print("\n" + "="*70 + "\n")


spearman_corr = df_selected_features.corr(method='spearman')
print("Spearman Rank Correlation Matrix:\n", spearman_corr)

plt.figure(figsize=(9, 7)) 
sns.heatmap(spearman_corr,
            annot=True,    
            cmap='viridis', 
            fmt=".2f",     
            linewidths=.8, 
            cbar_kws={'label': 'Spearman Correlation Coefficient'}) 
plt.title('Spearman Rank Correlation Heatmap of Pokémon Features')
plt.show()
```
