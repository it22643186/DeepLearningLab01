```python
import numpy as np
import pandas as pd
```

```python
 df = pd.read_csv('Au_nanoparticle_dataset.csv')
```

```python
# Question 1
Except_columns = df[['N_total', 'N_bulk', 'N_surface', 'R_avg']]

filtered_df = df.drop(columns=Except_columns, axis=1)

print("Successfully created filtered DataFrame:")
```

```python
# Question 2
filtered_df.head(1)
```

```python
print("\nMean values:")
print(Except_columns.mean())

print("\nStandard Deviation values:")
print(Except_columns.std())

print("\nQuartile values (25th, 50th/Median, 75th percentile):")
print(Except_columns.quantile([0.25, 0.5, 0.75]))
```

```python
import matplotlib.pyplot as plt
fig, axes = plt.subplots(nrows=1, ncols=len(Except_columns.columns), figsize=(20, 5))

for i, col in enumerate(Except_columns.columns):
    axes[i].hist(Except_columns[col], bins='auto') 
    axes[i].set_title(f'Histogram of {col}')
    axes[i].set_xlabel("Value")
    axes[i].set_ylabel("Frequency")

plt.show() 
```

```python
import seaborn as sns
```

```python
sns.pairplot(Except_columns)

plt.suptitle('Pairplot of Filtered Au Nanoparticle Dataset Features', y=1.02, fontsize=16) # Add a main title
plt.tight_layout(rect=[0, 0, 1, 0.98]) 


plt.show()
```

```python

g = sns.PairGrid(Except_columns)


g.map_diag(sns.histplot, kde=True, color='purple') 

g.map_lower(sns.kdeplot, cmap='Blues_d') 

g.map_upper(sns.histplot, cmap='Greens') 


g.fig.suptitle('Customized PairGrid: Histograms+KDE (Diag), Bivariate KDE (Lower), Bivariate Hist (Upper)',
               y=1.02, fontsize=16)

plt.show()

print("\n--- Customized PairGrid displayed. ---")
print("The diagonal plots now include both histograms and KDE curves.")
print("The lower triangle plots now show bivariate KDE contours.")
print("The upper triangle plots remain bivariate histograms.")
```
